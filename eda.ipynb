```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_path = 'Task_A/train.parquet'
val_path = 'Task_A/validation.parquet'
test_path = 'Task_A/test.parquet'
public_test_path = 'Task_A/test_sample.parquet'
```

## Train

```python
train_df = pd.read_parquet(train_path)
train_df.head()
```

```python
print(f"Number of training sampeles: {len(train_df):_}")
```

```python
sns.set_theme(style="white")

label_counts = train_df['label'].value_counts()

colors = sns.color_palette("pastel")

plt.figure(figsize=(6,6))
plt.pie(
    label_counts.values,
    labels=label_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors
)

plt.title("Label Distribution")
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8,5))
ax = sns.countplot(data=train_df, x='language')

# Add values on top of bars
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}', 
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom',
        fontsize=11
    )

plt.title("Label Distribution")
plt.xlabel("Label")
plt.ylabel("Count")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Prepare Data
train_df['dataset'] = 'Train'
public_test_df['dataset'] = 'Test'
combined_df = pd.concat([train_df, public_test_df])
combined_df['token_length'] = combined_df['code'].apply(lambda x: len(str(x).split()))

# 2. Vertical Visualization with Tight Zoom
sns.set_theme(style="whitegrid")

g = sns.displot(
    data=combined_df,
    x='token_length',
    hue='label',
    row='dataset',
    kind='hist',
    kde=True,
    element='step',
    palette='viridis',
    common_norm=False,
    height=4,
    aspect=2.5,
    facet_kws={'sharey': False},
    # Set bins manually to get better resolution in the skewed area
    bins=200 
)

# 3. Aggressive Zoom and Scale Fixes
for ax in g.axes.flat:
    # Zoom into where the action is (e.g., 0 to 800)
    # This makes the 512 line much more prominent
    ax.set_xlim(0, 800) 
    
    # ADDING LOG SCALE TO Y-AXIS (Optional)
    # Uncomment the line below if the bars are still too tall to see the KDE
    # ax.set_yscale('log') 

    # Red dashed line at 512
    ax.axvline(x=512, color='red', linestyle='--', linewidth=2)
    
    # Text annotation
    y_max = ax.get_ylim()[1]
    ax.text(520, y_max * 0.8, '512-token\nCut-off', color='red', fontweight='bold', fontsize=10)

# Labels
g.set_axis_labels("Number of Tokens (Zoomed)", "Count")
g.set_titles(row_template="{row_name} Set")
plt.subplots_adjust(top=0.9)
g.fig.suptitle("Code Token-length Distribution: Detailed View", fontsize=16)

plt.show()
```
